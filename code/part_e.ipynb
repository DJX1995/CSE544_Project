```python
import pandas as pd
import numpy as np
import math


ALdata = pd.read_csv('/content/AL_administered_SepNov.csv', index_col=0).to_numpy().squeeze()
AZdata = pd.read_csv('/content/AZ_administered_SepNov.csv', index_col=0).to_numpy().squeeze()

AL_mean = ALdata.mean()

AZ_mean = AZdata.mean()

D = ALdata - AZdata
D_mean = AL_mean - AZ_mean
D_variance = np.square(D - D_mean).sum() / (len(D) - 1)
D_se = np.sqrt(D_variance)

critical_value = 2.002
Test_stat = D_mean/(D_se/np.sqrt(len(D)))

print(f'|T| = {np.abs(Test_stat):.3f} and t score = {critical_value:3f}')
if np.abs(Test_stat) > critical_value:
  print('accept')
else:
  print('reject')
```
